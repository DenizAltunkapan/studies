```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.figure(figsize=(3,2),dpi=200) # Größe
plt.plot(x,y, label='2x', color="purple", linewidth = 3, marker='.', markersize=10, markerfacecolor= 'blue', linestyle='--') 
#betrachte shorthand notation

plt.title("Our first graph", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("x Axis")
plt.ylabel("y Axis")

#beschriftungen der achsen
plt.xticks([0,1,2,3,4,10]) 
plt.yticks([0,2,4,6,8,10])

plt.legend() #anzeige von 2x (siehe label bei plot)
plt.show()
```

```python
x2 = np.arange(0,4.5,0.5) #use numpy
plt.plot(x2, x2**2, 'r', label = 'x^2')
plt.legend()
plt.show()
```

### Bar Chart

```python
labels = ['A', 'B', 'C']
values = [1,4,2]

bars = plt.bar(labels, values)

bars[0].set_hatch('/') #schraffierung erster Balken
plt.figure(figsize=(2,1))

plt.show()
```

```python
gas= pd.read_csv('gas_prices.csv')
plt.plot(gas.Year, gas.USA, label='USA')
plt.plot(gas.Year, gas.Germany, label='Germany')
plt.title("Gas Prices in Usa vs. Germany")
plt.savefig('gasprice_figure.png',dpi=200)
plt.legend()
plt.show()
```

```python
def nicht_stetige_funktion(x):
    if x < 1:
        return 0.5
    elif x == 1:
        return np.nan  # Sprung
    else:  # x > 1
        return 1

# Array von x-Werten
x = np.linspace(0, 3, 1000)
y = np.array([nicht_stetige_funktion(i) for i in x])

plt.plot(x, y)
plt.title('Nicht-stetige Funktion')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.ylim(0, 1.5)

plt.xticks(np.arange(0, 4, 0.5))
plt.yticks(np.arange(0, 2, 0.5))
plt.grid()
plt.show()
```
