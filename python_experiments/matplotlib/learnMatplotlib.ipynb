```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### Basic Graph

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]
plt.figure(figsize=(3,2),dpi=200) # Größe
plt.plot(x,y, label='2x', color="purple", linewidth = 3, marker='.', markersize=10, markerfacecolor= 'blue', linestyle='--') 
#betrachte shorthand notation

plt.title("Our first graph", fontdict = {'fontname': 'Comic Sans MS', 'fontsize': 20})
plt.xlabel("x Axis")
plt.ylabel("y Axis")

#beschriftungen der achsen
plt.xticks([0,1,2,3,4,10]) 
plt.yticks([0,2,4,6,8,10])

plt.legend() #anzeige von 2x (siehe label bei plot)
plt.show()
```

```python
x2 = np.arange(0,4.5,0.5) #use numpy
plt.plot(x2, x2**2, 'r', label = 'x^2')
plt.legend()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def nicht_stetige_funktion(x):
    if x < 1:
        return 0.33
    elif (x > 0.95 and x<1.05) or (x>2.95 and x<3.05):
        return np.nan  # Sprung
    elif x > 1 and x < 3:
        return 0.83
    else:
        return 1

# Array von x-Werten
x = np.linspace(0, 4, 1000)
y = np.array([nicht_stetige_funktion(i) for i in x])

# Verwende plt.step statt plt.plot
plt.step(x, y, where='post', color='blue')
plt.title('1.4 Betrunkene Seeleute: Verteilungsfunktion')
plt.xlabel('x')
plt.ylabel('P(X<=x)')
plt.ylim(0, 1.25)

# Markiere die Sprungstellen explizit
plt.plot(0, 0.33, 'o', color='blue')
plt.plot(1, 0.33, 'o', color='blue', fillstyle='none')
plt.plot(1, 0.83, 'o', color='blue')
plt.plot(3, 0.83, 'o', color='blue', fillstyle='none')
plt.plot(3, 1, 'o', color='blue')

plt.xticks(np.arange(0, 5, 0.5))
plt.yticks(np.arange(0, 1.5, 0.25))
plt.grid()
plt.savefig('s1_a1.4_2.png')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Definieren der nicht-stetigen Verteilungsfunktion
def nicht_stetige_funktion(x):
    if x < 1:
        return 0
    elif (x > 0.95 and x < 1.05) or (x > 1.95 and x < 2.05) or (x > 2.95 and x < 3.05) or (x > 3.95 and x < 4.05) or (x > 4.95 and x < 5.05) or (x > 5.95 and x < 6.05):
        return np.nan  # Sprungstelle
    elif x > 1 and x < 2:
        return 0.00000002
    elif x > 2 and x < 3:
        return 0.00002
    elif x > 3 and x < 4:
        return 0.0001
    elif x > 4 and x < 5:
        return 0.018
    elif x > 5 and x < 6:
        return 0.162
    else:
        return 1

# Array von x-Werten für die Berechnung
x = np.linspace(0, 9, 1000)
y = np.array([nicht_stetige_funktion(i) for i in x])

# Zeichnen der Treppenfunktion
plt.step(x, y, where='post', color='blue')
plt.title('1.6 Würfelmaximum: Verteilungsfunktion')
plt.xlabel('x')
plt.ylabel('P(X <= x)')
plt.ylim(0, 1.1)

# Manuelles Setzen der Sprungstellen und zugehöriger Werte
sprungstellen_x = [1, 2, 3, 4, 5, 6]
# Werte an den Sprungstellen
werte_vor_sprung = [0.0, 0.0001, 0.001, 0.001, 0.018, 0.162]
werte_nach_sprung = [0.00000002, 0.0001, 0.001, 0.018, 0.162, 1]

# Markiere die Sprungstellen explizit
for i in range(len(sprungstellen_x)):
    # Gefüllter Punkt vor dem Sprung
    plt.plot(sprungstellen_x[i], werte_vor_sprung[i], 'o', color='blue', fillstyle='none')
    # Leerer Punkt nach dem Sprung
    if np.isfinite(werte_nach_sprung[i]):
        plt.plot(sprungstellen_x[i], werte_nach_sprung[i], 'o', color='blue')

# Feinabstimmung der Achsen und Gitter hinzufügen
plt.xticks(np.arange(0, 9, 1))
plt.yticks(np.arange(0, 1.1, 0.05))
plt.grid(True)
plt.savefig('s1_a1.6.png')
# Plot anzeigen
plt.show()
```
